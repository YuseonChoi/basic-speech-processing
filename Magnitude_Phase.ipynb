```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import math
```

```python
x = torch.linspace(-10, 10, 500)
y = torch.cos(2*math.pi/5*x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
Y = np.fft.fft(y)
# function
print(f"magnitude: {np.abs(Y)[:5]}\n {np.abs(Y).shape}\n")
print(f"angle: {np.angle(Y)[:5]}\n {np.angle(Y).shape}\n")

Y_m = np.sqrt((Y.real)**2+(Y.imag)**2)
Y_p = np.arctan(Y.imag/Y.real)
# implementation
print(f"magnitude: {Y_m[:5]}\n {Y_m.shape}\n")
print(f"angle: {Y_p[:5]}\n {Y_p.shape}\n")
```
